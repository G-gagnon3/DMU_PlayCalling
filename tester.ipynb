```python
import nfldat as nd
import pandas as pd
```

```python
dat = nd.importer.read_span(2021, 2023)
```

```python
#dat.to_csv('2010_to_2018.csv', index=False)
sim = nd.simulate.sim(dat, nd.transition.transitionModelStandard)
```

## Scenario 1: 4th & 10 @ 30

```python
drive_start, down, distance, sim_range = (30, 4, 10, 2)
sim.choose_4th(drive_start, distance, sim_range, 30)
```

## Scenario 2: 4th & 10 @ 70

```python
drive_start, down, distance, sim_range = (70, 4, 10, 2)
sim.choose_4th(drive_start, distance, sim_range, 30)
```

## Scenario 3: 3rd & 10 @ 60

```python
drive_start, down, distance, sim_range = (60, 3, 10, 2)
sim.choose_other(drive_start, down, distance, sim_range, 30)
```

## Scenario 4: 1st & 10 @ 10

```python
drive_start, down, distance, sim_range = (10, 1, 10, 2)
sim.choose_other(drive_start, down, distance, sim_range, 30)
```

## Scenario 5: 4th & Goal @ 1

```python
drive_start, down, distance, sim_range = (1, 4, 1, 2)
sim.choose_4th(drive_start, distance, sim_range, 30)
```

```python
drive_starts = [20, 30, 40]
drive_starts = [70]
drive_starts = [10, 30, 50, 70, 90]
down = 1
distance = 60
sim_range = 2

sims_per_spot = 20
epas = [0 for i in range(sims_per_spot)]
for j, drive_start in enumerate(drive_starts):
    cur_reward = 0
    for i in range(sims_per_spot):
        print(f"Sim {i}")
        play = sim.choose_other(drive_start, down, distance, sim_range, 30)
        reward, sequence = sim.simulate_drive(play, down, distance, drive_start, sim_range, "optimal", 40)
        print(reward, sequence)
        cur_reward += reward
    cur_reward = cur_reward/sims_per_spot
    epas[j] = cur_reward
epa_dict = {}
for i, drive_start in enumerate(drive_starts):
    epa_dict[drive_start] = (epas[i], nd.simulate.get_epa(1,drive_start))
print(epa_dict)

```
